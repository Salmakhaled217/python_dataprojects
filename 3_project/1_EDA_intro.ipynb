```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda skill: ast.literal_eval(skill)if pd.notna(skill)else skill)
```

```python
df_DA_US=df[(df['job_title_short']=='Data Analyst')&(df['job_country']=='United States')]
```

```python
df_DA_US
```

```python
df_plot=df_DA_US['job_location'].value_counts().head(10).to_frame()
```

```python
import seaborn as sns
```

```python
sns.set_theme( style='ticks')
sns.barplot(data=df_plot,x='count',y='job_location',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('number of jobs for data analyst in the united states')
plt.xlabel('count of jobs')
plt.ylabel('')
plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(), labels=['True', 'False'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.suptitle('Benefit Analysis of Data Jobs', fontsize=16)
plt.legend(loc='upper center',bbox_to_anchor=(-0.55,0),ncol=2)
plt.show()
```

```python
df_plot=df_DA_US['company_name'].value_counts().head(10).to_frame()

sns.set_theme( style='ticks')
sns.barplot(data=df_plot,x='count',y='company_name',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('number of companies for data analyst in the united states')
plt.xlabel('count of jobs')
plt.ylabel('')
plt.show()
```
