```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_usa=df[df['job_country']=='United States'].copy()
df_usa['job_posted_month']=df_usa['job_posted_date'].dt.strftime('%B')
df_us_pivot=df_usa.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month']=pd.to_datetime(df_us_pivot['job_posted_month'],format='%B').dt.month
df_us_pivot.sort_values('month',inplace=True)
df_us_pivot.set_index('job_posted_month',inplace=True)
df_us_pivot.drop(columns='month',inplace=True)
```

```python
software_jobs=pd.read_csv(r"C:\Users\salma\Downloads\software_jobs.csv",index_col='job_posted_month')
```

```python
df_us_merged=df_us_pivot.merge(software_jobs,on='job_posted_month')
```

```python
top_5=df_us_merged.sum().sort_values(ascending=False).head(5).index.to_list()
top_5
```

```python
df_us_merged[top_5].plot(kind='line')
plt.ylabel('job count')
plt.xlabel('2023')
plt.title('job posting for top five tech jobs')
plt.legend()
plt.show()
```
