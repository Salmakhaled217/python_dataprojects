```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_usa=df[df['job_country']=='United States'].copy()
```

```python
df_usa['job_posted_month']=df_usa['job_posted_date'].dt.strftime('%B')
```

```python
df_usa
```

```python
df_us_pivot=df_usa.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_us_pivot
```

```python
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month']=pd.to_datetime(df_us_pivot['job_posted_month'],format='%B').dt.month
df_us_pivot.sort_values('month',inplace=True)
df_us_pivot.set_index('job_posted_month',inplace=True)
df_us_pivot.drop(columns='month',inplace=True)
df_us_pivot
```

```python
df_us_pivot.plot(kind='line')
```

```python
top_3=df_usa['job_title_short'].value_counts().head(3)
top_3=top_3.index.to_list()
```

```python
df_us_pivot[top_3].plot(kind='line')
plt.xlabel('2023')
plt.ylabel('job count')
plt.title('Monthly job postong for top data jobs in us')
plt.show()
```
