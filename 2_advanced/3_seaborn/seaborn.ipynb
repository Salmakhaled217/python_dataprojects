```python
import seaborn  as sns
import pandas as pd
import numpy as np
```

```python
penguins=sns.load_dataset('penguins')
```

```python
sns.set_context('notebook')
sns.histplot(data=penguins,x='body_mass_g',hue='sex',multiple='stack')
```

```python
sns.displot(data=penguins,x='body_mass_g',hue='sex',kind='kde',fill=True)
```

```python
sns.histplot(data=penguins,x='body_mass_g',hue='sex',multiple='stack',kde=True)
```

```python
sns.set_theme(style='ticks')
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='sex',palette='Set2')
sns.set_context('notebook')
sns.despine()
```

```python
sns.set_theme(style='whitegrid')
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='sex')
sns.despine(left=True)
```

```python
sns.scatterplot(data=penguins,x='species',y='body_mass_g',hue='island',palette='Set1',style='sex')
```

```python
sns.stripplot(data=penguins,x='species',y='body_mass_g',hue='island',jitter=True,dodge=True)
```

```python
sns.swarmplot(data=penguins,x='species',y='body_mass_g',hue='island')
```

```python
sns.regplot(data=penguins,x='flipper_length_mm',y='body_mass_g',color='green',scatter=False)
```

```python
sns.lineplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='island',style='sex')
```

```python
sns.jointplot(data=penguins,x='flipper_length_mm',y='body_mass_g')
```

```python
sns.jointplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='sex',kind='kde')
```

```python
sns.barplot(data=penguins,x='species',y='body_mass_g',hue='sex',palette=['pink','skyblue'],estimator=np.sum)
```

```python
sns.countplot(data=penguins,x='species',hue='island',palette="Set2")
```

```python
sns.boxplot(data=penguins,x='species',hue='island',y='body_mass_g')
```

```python
sns.violinplot(data=penguins,x='species',hue='island',y='body_mass_g',palette="Set2",inner='box')
```

```python
sns.violinplot(data=penguins,x='species',y='body_mass_g',palette="Set2")
sns.swarmplot(data=penguins,x='species',y='body_mass_g',color='black',size=3)
```

```python
sns.kdeplot(data=penguins,x='body_mass_g',hue='island',fill=True)
```

```python
columns=['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
sns.heatmap(data=penguins[columns].corr(),annot=True,cmap='Blues',vmin=-.2,linewidth="2",linecolor='blue')
```

```python
sns.rugplot(data=penguins,y='body_mass_g',hue='species',palette="Set2",height=.2)
```

```python
sns.pairplot(data=penguins,hue='sex',palette="Set2",diag_kind="kde")
```

```python
graph=sns.PairGrid(data=penguins,hue='sex',palette="Set2")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
```
